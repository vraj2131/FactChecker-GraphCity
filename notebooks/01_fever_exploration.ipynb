# FEVER Exploration

Goals:

```python
from pathlib import Path

import pandas as pd
```

```python
CLAIMS_PATH = Path("../data/processed/fever_claims.parquet")
EVIDENCE_PATH = Path("../data/processed/fever_evidence.parquet")

CLAIMS_SAMPLE_PATH = Path("../data/processed/fever_claims_sample.parquet")
EVIDENCE_SAMPLE_PATH = Path("../data/processed/fever_evidence_sample.parquet")
```

```python
claims_df = pd.read_parquet(CLAIMS_PATH)
evidence_df = pd.read_parquet(EVIDENCE_PATH)

claims_sample_df = pd.read_parquet(CLAIMS_SAMPLE_PATH)
evidence_sample_df = pd.read_parquet(EVIDENCE_SAMPLE_PATH)

print("claims_df shape:", claims_df.shape)
print("evidence_df shape:", evidence_df.shape)
print("claims_sample_df shape:", claims_sample_df.shape)
print("evidence_sample_df shape:", evidence_sample_df.shape)
```

Sample SUPPORTS evidence rows

```python
supports_rows = evidence_df[evidence_df["label"] == "SUPPORTS"]
supports_rows.head(10)
```

Sample REFUTES evidence rows

```python
refutes_rows = evidence_df[evidence_df["label"] == "REFUTES"]
refutes_rows.head(10)
```

Inspect one SUPPORTS claim with its evidence rows

```python
support_claim_id = supports_rows.iloc[0]["claim_id"]
support_claim = claims_df[claims_df["claim_id"] == support_claim_id]
support_evidence = evidence_df[evidence_df["claim_id"] == support_claim_id]

print("SUPPORT CLAIM")
display(support_claim)

print("\nSUPPORT EVIDENCE")
display(support_evidence)
```

Inspect one REFUTES claim with its evidence rows

```python
refute_claim_id = refutes_rows.iloc[0]["claim_id"]
refute_claim = claims_df[claims_df["claim_id"] == refute_claim_id]
refute_evidence = evidence_df[evidence_df["claim_id"] == refute_claim_id]

print("REFUTE CLAIM")
display(refute_claim)

print("\nREFUTE EVIDENCE")
display(refute_evidence)
```

Inspect one NOT ENOUGH INFO claim

```python
nei_rows = evidence_df[evidence_df["label"] == "NOT ENOUGH INFO"]
nei_rows.head(10)
```

```python
nei_claim_id = nei_rows.iloc[0]["claim_id"]
nei_claim = claims_df[claims_df["claim_id"] == nei_claim_id]
nei_evidence = evidence_df[evidence_df["claim_id"] == nei_claim_id]

print("NEI CLAIM")
display(nei_claim)

print("\nNEI EVIDENCE")
display(nei_evidence)
```

Useful debugging query by claim id

```python
def inspect_claim(claim_id: int):
    claim_row = claims_df[claims_df["claim_id"] == claim_id]
    evidence_rows = evidence_df[evidence_df["claim_id"] == claim_id]

    print("CLAIM ROW")
    display(claim_row)

    print("\nEVIDENCE ROWS")
    display(evidence_rows)
```
